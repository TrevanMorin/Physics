```python
import numpy as np
import matplotlib.pyplot as plt

# Charging data
t_charging = np.array([22.36, 22.04, 22.16, 21.6, 21.58])  # Time (s)
V_charging = np.array([7.21112, 7.19888, 7.21112, 7.19888, 7.21112])  # Voltage (V)

# Discharging data
t_discharging = np.array([22.36, 21.83, 22.24, 22.46, 22.52])  # Time (s)
V_discharging = np.array([4.19888, 4.21112, 4.19888, 4.21112, 4.19888])  # Voltage (V)

# Create Charging Plot
plt.figure()
plt.plot(t_charging, V_charging, marker='o', linestyle='-', color='blue')
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Capacitor Charging Curve")
plt.grid()
plt.show()

# Create Discharging Plot
plt.figure()
plt.plot(t_discharging, V_discharging, marker='o', linestyle='-', color='red')
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Capacitor Discharging Curve")
plt.grid()
plt.show()
```
